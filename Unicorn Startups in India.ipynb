**1.Which Startup has the highest valuation**

```python
Unicorn.head()
```

```python
SortedUnicorn = Unicorn.sort_values(by="Valuation ($B)",ascending=False)
```

```python
SortedUnicorn.head()
SortedCompanies = SortedUnicorn["Company"]
SortedValuation = SortedUnicorn["Valuation ($B)"]
```

**Plotting the Bar graph for the unicorns**

```python
plt.figure(figsize=(18,7))
plt.bar(SortedCompanies,SortedValuation)
plt.xlabel("Unicorn Companies")
plt.xticks(SortedCompanies,rotation="vertical",size=9)
plt.ylabel("Valuation in Billions")
plt.grid()
plt.show()
```

**2. Which month has the highest number of startups**

```python
Unicorn.head()
```

```python
Unicorn["Month"] = Unicorn["Entry"].apply(lambda x : x.split("-")[1])
```

```python
CountofValues = Unicorn["Month"].value_counts()
```

```python
dictionary = dict(CountofValues)
```

```python
Months = dictionary.keys()
Counts = dictionary.values()
```

```python
plt.bar(Months,Counts)
plt.xticks(list(Months),rotation="vertical",size=8)
plt.xlabel("Months")
plt.ylabel("No of unicorns")
plt.show()
```

**Question 4: Firms that invested the most**

```python
inv = []
for i in Unicorn["Select Investors"]:
    investors = i.split(",")
    for every in investors:
        inv.append(every.strip())
inv = pd.Index(inv)
plt.title("Companies invested by Investors")
inv.value_counts().head(10).plot(kind='bar')
plt.xlabel("Number of Startups invested")
plt.ylabel("Investors")
plt.show()
```

**Question 5: States which has the maximum number of startups**

```python
Locat = []
for i in Unicorn["Location"]:
    investors = i.split("/")
    for every in investors:
        Locat.append(every.strip())
Locat = pd.Index(Locat)
plt.figure(figsize=(17,10))
plt.title("Startups in Various Locations")
Locat.value_counts().plot(kind='barh')
plt.xlabel("Number of companies")
plt.ylabel("Location")
plt.grid()
plt.show()
```
